# **FILE FOR CREATING A DATASET FOR LOGISTIC REGRESSION**

```python
#Import required libraries
import pandas as pd
import numpy as np
```

```python
# Access folder containing acitivity data
data_folder = "/content/drive/Shareddrives/OSM-ODECO/HOTOSM/dataforanalysis/"
in_drive = True  # True to mount a drive while working in Google Colab
if in_drive:
    from google.colab import drive
    drive.mount("/content/drive", force_remount=True)
```

```python
# Read states data
total=pd.read_csv(data_folder +"splits_invalidations.csv")
total['projectId'] = total['taskId'].str.split('_').str[0]
total['task'] = total['taskId'].str.split('_').str[1]
total['projectId']=total['projectId'].astype(str).astype(int)
total['task']=total['task'].astype(str).astype(int)
listinvalidated=total[total['action']=="INVALIDATED"].taskId.unique().tolist()
total['invalidated'] = np.where(total['taskId'].isin(listinvalidated), 1, 0)
total["start"]=pd.to_datetime(total["start"],infer_datetime_format='mixed',errors='coerce')
total["complete"]=pd.to_datetime(total["complete"],infer_datetime_format='mixed',errors='coerce')
total["duration"]=(total.complete-total.start).astype('timedelta64[s]')
total
```

```python
# Read project data
projects=pd.read_excel(data_folder +"projects.xlsx")
total=total.merge(projects, on="projectId")
projects
```

```python
# Read building task area
area=pd.read_csv(data_folder +"regression.csv")
area=area[["taskId","area_sqm",	"percentage_area_covered_by_building"]]
area["areamapped"]=round(area["area_sqm"]*area["percentage_area_covered_by_building"]/100,0)
area["areamapped"] = np.where(area["areamapped"] < 0, 0, area["areamapped"])
total=total.merge(area, on="taskId")
```

```python
# Keep states before first validation
total=total.sort_values(["taskId","start"]).reset_index()
firstvalidation=total[total['action'].isin(['VALIDATED','INVALIDATED'])]
firstvalidation=firstvalidation.drop_duplicates(subset=['taskId'])
firstvalidation=firstvalidation[['taskId','start']]
firstvalidation.columns=['taskId','validationtime']
total=total.merge(firstvalidation[['taskId','validationtime']], on="taskId")
firstvalidation
```

```python
# Identify task done by more than one mapper
total=total[total['start'] <= total['validationtime']]
groups = pd.DataFrame(total[total['action'].isin(['LOCKED_FOR_MAPPING','MAPPED','AUTO_UNLOCKED_FOR_MAPPING','SPLIT','BADIMAGERY'])].groupby('taskId')['actionBy'].nunique()).reset_index()
groups.columns = ['taskId','mappers']
total=total.merge(groups[["taskId","mappers"]], on="taskId")
total
```

```python
# Identify tasks with splits, autounlocked for mapping, autounlocked for validation to remove them.
# Identify tasks with locked for mapping and locked for validation
listsplit=total[total['action']=="SPLIT"].taskId.unique().tolist()
listautounlockedformapping=listsplit=total[total['action']=="AUTO_UNLOCKED_FOR_MAPPING"].taskId.unique().tolist()
listautounlockedforvalidation=listsplit=total[total['action']=="AUTO_UNLOCKED_FOR_VALIDATION"].taskId.unique().tolist()
listsmapping=total[total['action']=="LOCKED_FOR_MAPPING"].taskId.unique().tolist()
listsvalidation=total[total['action']=="LOCKED_FOR_VALIDATION"].taskId.unique().tolist()
total=total[total["taskId"].isin(listsmapping)]
total=total[total["taskId"].isin(listsvalidation)]
total=total[~total["taskId"].isin(listsplit)]
total=total[~total["taskId"].isin(listautounlockedformapping)]
total=total[~total["taskId"].isin(listautounlockedforvalidation)]
```

```python
# Identify the mapping level of the contributor who declared the task as mapped
whomapped=total[total["action"]=="LOCKED_FOR_MAPPING"].drop_duplicates(subset=['taskId'],keep="last")
whomapped=whomapped[["taskId","mappingLevel"]]
whomapped.columns=["taskId","whomapped"]
total=total.merge(whomapped[["taskId","whomapped"]], on="taskId")
total
```

```python
# Calculate total mapping time per task
timemapping = pd.DataFrame(total[total["action"]=="LOCKED_FOR_MAPPING"].groupby("taskId")["duration"].sum()).reset_index()
timemapping.columns = ["taskId","durationmapping"]
# Calculate total validation time per task
timevalidation = pd.DataFrame(total[total["action"]=="LOCKED_FOR_VALIDATION"].groupby("taskId")["duration"].sum()).reset_index()
timevalidation.columns = ["taskId","durationvalidation"]
# Calculate relative duration of taks
durations=total.drop_duplicates(subset=['taskId']).merge(timemapping, on="taskId")
durations=durations.merge(timevalidation, on="taskId")
durations = durations[['taskId','durationmapping',"durationvalidation","mappers","areamapped"]]
durations['group'] = np.where(durations['mappers'] > 1, 1, 0)
durations['logareamapped'] = np.log(durations['areamapped']+1)
durations['durationmapping'] = pd.to_numeric(durations.durationmapping)
durations['durationvalidation'] = pd.to_numeric(durations.durationvalidation)
durations
```

```python
# Create dataframe for regression
total=total[["taskId","priority","difficulty","areamapped","mappers","invalidated","projectId","whomapped"]]
regression=total.drop_duplicates(subset=['taskId'])
regression['group'] = np.where(regression['mappers'] > 1, 1, 0)
regression['areamapped'] = np.log(abs(regression['areamapped'])+1)
regression['projectId'] = regression['projectId'].map(str)
regression=regression.dropna()
regression['priority'] = regression['priority'].map({'LOW': '1. LOW', 'MEDIUM': '2. MEDIUM', 'HIGH': '3. HIGH', 'URGENT': '4. URGENT'})
regression['difficulty'] = regression['difficulty'].map({'EASY': '1. EASY', 'MODERATE': '2. MODERATE', 'CHALLENGING': '3. CHALLENGING'})
regression['whomapped'] = regression['whomapped'].map({'BEGINNER': '1. BEGINNER', 'INTERMEDIATE': '2. INTERMEDIATE', 'ADVANCED': '3. ADVANCED'})
regression=regression.merge(durations[["taskId","durationmapping","durationvalidation"]], on="taskId")
regression['durationvalidationper'] = regression['durationvalidation']/(regression['durationvalidation']+regression['durationmapping'])*100
regression['qareamapped'] = pd.qcut(regression['areamapped'], 2, labels=["Q1","Q2"])
regression['qdurationvalidationper'] = pd.qcut(regression['durationvalidationper'], 2, labels=["Q1","Q2"])
regression
```
